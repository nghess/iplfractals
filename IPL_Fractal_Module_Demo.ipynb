#Generate a 2d fractal
We create a fractal using the Generate() function. We can define beta, seed, size and whether the fractal is 2d or 3d.

```python
static = fractal.Generate(beta=3, seed=3131, size=512, dimension=2)
```

###Preview the fractal
Once we've create a fractal object, we can run various methods on it to get previews, box counts, or spectral slope.

```python
static.preview2d()
```

###Get D with box counting

```python
static.boxcount()
```

#Generate a 3d fractal
To generate a 3d fractal, we again use the Generate() function, changing the dimension argument from 2 to 3

```python
dynamic = fractal.Generate(beta=4, seed=1515, size=256, dimension=3)
```

###Preview 3d fractals as 2d frames
For 3d fractals, the preview2d() function grabs the top frame in the stack by default. 
It also takes an argument to specify other frames on a 1-99 scale. An argument of 50 will find the middle frame in the stack.

```python
#First frame in the stack
dynamic.preview2d()
```

```python
#50% of the way through the stack
dynamic.preview2d(50)
```

###Preview 3d fractals as volumes
We can also preview 3d fractals as cubic volumes. This method is computationally intensive at high resolutions, so as a time saver, I automatically resize the fractal to 64 voxels per side before rendering.

```python
dynamic.preview3d()
```

###Preview 3d fractals as animations
In the offline version of this library, it's possible to get animated previews with the previewAnim() function. However, Colab doesn't have good support for animation, so I'm omitting that for today.

###Get D with box counting
The boxcount() function automatically detects that a fractal is 3d and produces a cube-counted D.

```python
dynamic.boxcount()
```

###Average box counting D
For 3d fractals, we have the option to analyze each slice individually, then average those values. This function is somewhat slow, but it's a nice validation measure for the 3d boxcount. We see that the numbers agree nicely.

```python
dynamic.avgBoxcount()
```
